# Понятие корреляции
## Ковариация и корреляция

`Ковариация` - это **линейная мера**, показывающая степень изменчивости двух случайных величин в тандеме.

$$ cov(X, Y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{x})}}{N - 1} $$
где N - количество случайных величин, а единица - количество степеней свободы.

Однако только по **абсолютному** значению ковариации нельзя судить о том, насколько сильно величины взаимосвязаны, так как масштаб ковариации зависит от их дисперсий.  
Значение ковариации можно нормировать, поделив её на произведение среднеквадратических отклонений (квадратных корней из дисперсий) случайных величин (иначе говоря, деля на ${\sigma_x\sigma_y}$, мы загоняем в рамки [-1, 1]).

Полученная величина называется `коэффициентом корреляции Пирсона`, который всегда находится в интервале от −1 до 1:

$$ r(x, y) = \frac{cov(x, y)}{\sigma_x\sigma_y}$$

`Корреляция` - это особый случай ковариации, который может быть получен при стандартизации данных.  
Теперь, когда дело доходит до выбора, который является лучшей мерой взаимосвязи между двумя переменными, корреляция предпочтительнее ковариации, поскольку она не зависит от изменения местоположения и масштаба, а также может использоваться для сравнения между две пары переменных.

[Источник](https://ru.gadget-info.com/difference-between-covariance)  | КО-ВАРИАЦИЯ | КО-РРЕЛЯЦИЯ
------------- | ------------- | -----------
Определение  | `Ковариация` - это **линейная мера**, показывающая степень изменчивости двух случайных величин в тандеме  | `Корреляция` - это **статистическая мера**, которая показывает, насколько сильно связаны две переменные.
Что это?  | Мера корреляции  | Масштабная версия ковариации
Ценности  | Лежат между -∞ и + ∞  | Лежат между -1 и +1
Изменение масштаба | Влияет на ковариацию  | Не влияет на корреляцию
Единица измерения  | Нет  | Да

## Вывод формулы корреляции

Давайте остановимся на формуле коэффициента корреляции, которую мы получили:
$$ r(x, y) = \frac{cov(x, y)}{\sigma_x\sigma_y}$$
запишем формулу чуть подробнее и выполним возможные преобразования:

$$ r(x, y) = \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{(N - 1)\sqrt{\sum{\frac{(x_i - \bar{X})^2}{N-1}}}\sqrt{\sum{\frac{(y_i - \bar{Y})^2}{N-1}}}} $$

теперь вынесем $\frac{1}{(N - 1)}$ из под корней:

$$ r(x, y) = \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{(N - 1)\frac{1}{(N-1)}\sqrt{\sum{(x_i - \bar{X})^2}}\sqrt{\sum{(y_i - \bar{Y})^2}}} $$

и сократим $(N - 1)$:

$$ r(x, y) = \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{\sqrt{\sum{(x_i - \bar{X})^2}}\sqrt{\sum{(y_i - \bar{Y})^2}}} $$

таким образом, мы сократили $N - 1$ в знаменателе и получили финальную формулу для коэффициента корреляции, которую вы часто сможете встретить в учебниках:

$$ r(x, y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{Y})}}{\sqrt{\sum{(x_i - \bar{x})^2}\sum{(y_i - \bar{Y})^2}}} $$

### Альтернативный взгляд "под углом косинуса"
Если представить, что все $x$-ы мы загоним в матрицу $X$, все $y$-ки мы загоним в матрицу $Y$. Каждому значению $x$-a соответствет значение $y$-a:
$$X= \begin{bmatrix}x_{1}\\x_{2}\\...\\x_{n}\end{bmatrix}, Y= \begin{bmatrix}y_{1}\\y_{2}\\...\\y_{n}\end{bmatrix}$$

Тогда мы можем сравнить эти два вектора. Они указывают на какие-то точки в многомерном пространстве.

Теперь нам хочется посмотреть, а на сколько они совпадают и самый простой способ это постмотреть какой между ними угол. Для этого нам понадобиться [скалярное произведение](https://ru.wikipedia.org/wiki/%D0%A1%D0%BA%D0%B0%D0%BB%D1%8F%D1%80%D0%BD%D0%BE%D0%B5_%D0%BF%D1%80%D0%BE%D0%B8%D0%B7%D0%B2%D0%B5%D0%B4%D0%B5%D0%BD%D0%B8%D0%B5):
$$X \cdot Y=|X||Y| \cos{(\alpha)} \Rightarrow \cos{(\alpha)}=\frac{X \cdot Y}{|X||Y|}$$
Где:
$$X \cdot Y= \sum_{i=0}^{n} x_i y_i \:\: ,\text{а} \:\: |X|=\sqrt{\sum_{i=0}^{n}x_i^2} \:\: \text{и} \:\: |Y|=\sqrt{\sum_{i=0}^{n}y_i^2}$$
Вы могли уже догадаться, что: 
$$x_i=x_{i-\text{our}}-\bar{X} \:\:\text{и}\:\: y_i=y_{i-\text{our}}-\bar{Y}$$
Тоесть, если всё это собрать вместе, то в результате получается почти наша формула из видео (и точно формула из следующего стэпа). Эта штука называется [`косинусным расстоянием(сходстовом)`](https://ru.wikipedia.org/wiki/%D0%92%D0%B5%D0%BA%D1%82%D0%BE%D1%80%D0%BD%D0%B0%D1%8F_%D0%BC%D0%BE%D0%B4%D0%B5%D0%BB%D1%8C):
$$similarity=\cos{(\theta)}=\frac{A \cdot B}{||A||||B||}=\frac{\sum_{i=1}^{n}A_i \times B_i}{\sqrt{\sum_{i=1}^{n}(A_i)^2} \times \sqrt{\sum_{i=1}^{n}(B_i)^2}}$$

## Пример
Демонстрация работы ковариации и корреляции

```python
def cov(x, y):
    assert x.size == y.size
    return ((x - x.mean()) * (y - y.mean())).sum()/(x.size - 1)

def cor(x, y):
    return cov(x, y)/(np.std(x, ddof=1)*np.std(y, ddof=1))

# функция имитирущая случаные факторы
# р - настолько существенным будет случайный фактор
def randomize(arr, p):
    alpha = np.max(arr) - np.min(arr)
    res = np.zeros(arr.shape)
    for i, v in enumerate(arr):
        sign = 1 if r.choice([True, False]) else -1
        res[i] = v + sign*alpha*r.random()*p
    return res
```

# Регрессия с одной независимой переменной

<center><img src="slides/week_3/week_3-14.png" title="slide 14" width="50%" height="50%"/></center>

`Регрессионный анализ` - набор статистических методов исследования влияния одной или нескольких независимых переменных $X_{1}, X_{2} , ... , X_{p}$ на зависимую переменную $Y$.  
Независимые переменные иначе называют `регрессорами` или `предикторами`, а зависимые переменные — `критериальными`. Терминология зависимых и независимых переменных отражает лишь математическую зависимость переменных (см. Корреляция), а не причинно-следственные отношения.  
Наиболее распространенный вид регрессионного анализа — `линейная регрессия` (*linear regression*), когда находят линейную функцию, которая, согласно определённым математическим критериям, наиболее соответствует данным.  
Например, в методе наименьших квадратов вычисляется прямая(или гиперплоскость), сумма квадратов между которой и данными минимальна.

$$ y = b_0 + b_1x $$
$b_0$ (intercept) - значение пересечения линии с осью Y  
$b_1$ (slope) - задаёт наклон линии регрессии

## Метод наименьших квадратов

<center><img src="slides/week_3/week_3-15.png" title="slide 15" width="75%" height="75%"/></center>

`Метод наименьших квадратов` (МНК, *Least Squares*, *LS*) - это способ нахождения оптимальных параметров линейной регресссии ($b_0$,  $b_1$), таких, что сумма квадратов ошибок (остатков) была минимальная. Расчёт параметров выполняется по формулам:
$$ b_0 = \bar{Y} - b_1\bar{X} \:\:, \:\: b_1 = \frac{sd_y}{sd_x}r_{xy}$$
где $sd$ - стандартное отклонение, а $r$ - коэффициент корреляции.

Сущность МНК (обычного, классического) заключается в том, чтобы найти такие параметры $b$, при которых **сумма квадратов отклонений** (**ошибок**, для регрессионных моделей их часто называют **остатками регрессии**) $e_{t}$ будет минимальной

```python
'''Демонстрация МНК'''
N = 100

X = np.array([np.random.randint(50,100) for i in range(N)])
Y = np.array([i+np.random.randint(80,120)-np.random.randint(-20,20) for i in X])

plt.scatter(X, Y)
b1, b0 = np.polyfit(X, Y, 1) #  b0 - intercept, b1 - slope
plt.plot(X, b0 + b1*X, color='red')
print("b0={}\nb1={}".format(b0,b1))
```

# Гипотеза о значимости взаимосвязи и коэффициент детерминации

<center><img src="slides/week_3/week_3-16.png" title="slide 16" width="75%" height="75%"/></center>

Если мы хотим ответить на вопрос: а насколько статистически значима взаимосвязь двух наших переменных?  
Нулевая гипотеза $H_0:b_1=0$, альтернативная гипотеза $H_1:b_1 \ne 0$
$$t=\frac{b_1-0}{se}=\frac{b_1}{se} \:\:,\:\: df=N-2$$
**Основная идея:** `t-критерий` проверяет гипотезу о том, что наш коэффициент $b_1$, ответственный за угол наклона, отличен от нуля в генеральной совокупности.

## Коэффициент детерминации

<center><img src="slides/week_3/week_3-17.png" title="slide 17" width="75%" height="75%"/></center>

$R^2$ - доля дисперсии зависимой переменной ($Y$), объясняемая регресивной моделью.  
Квадрат коэффициента корреляции ($r$) выборки, как правило, обозначается  и называется **коэффициентом детерминации**.

К примеру, как **качетсво** определеяется **ценой** нашего товара. На левом рисунке синим обозначена регрессионная прямая, которая совсем чуть-чуть наклонена вверх, потому что коэффициент кореляции $r=0.04$ слишком мал. И мы хотим ответить на вопрос: **в какой мере изменчивать переменной качетсво объясняется взаимосвязью с переменной цена**? 

$$R^2=\frac{SS_{res}}{SS_{total}}$$

$SS_{res}$ - расстояние от нашего наблюдения до регрессионной прямой, возведенной в квадрат.

На правом рисунке мы видим явно выраженную корреляцию ($r=0.99 \rightarrow 1$). Наши точки очень близко расположены с регрессионной прямой и сумма квадратов остатков ($SS_{res}$) будет гораздо меньше общей суммы квадратов. Это будет приводить к тому, что отношение ($\frac{SS_{res}}{SS_{total}}$) будет становиться всё меньше и меньше, чем лучше наша регрессионная модель.  
В нашем случае, $R^2 \approx 1$ говорит о том, что практически 100% изменчивости нашей зависимой переменной ($Y$) обуславливается ее взаимосвязью с нашей независимой переменной ($X$). Т.е. наша регрессионная модель очень хорошо объясняет и описывает поведение и изменчивость зависимой переменной ($Y$). Поэтому, чем больше $R^2$, тем лучше наша модель справляется с поставленной задачей, а именно объясняет и предсказывает значения зависимой переменной.

**Предиктор** (*predictor*,  «предсказатель»)- это прогностический параметр; средство прогнозирования.

1. Линейная взаимосвязь $X$ и $Y$. Тут оченвидно, потому что какой смысл это делать, если переменные распределенны нелинейным оборазом.
2. Нормальное распределение остатков.
3. Гомоскедастичность - постоянная изменчивость остатков на всех уровнях независимой переменной.

Воспользуемся [сервисом](https://gallery.shinyapps.io/slr_diag/).

Дополнительно из комментариев:  
**Корреляционная прямая** показывает направление (положительное или отрицательное) и "силу" взаимосвязи. Но в полной мере она входные данные не описывает.   
**Регрессионная прямая** описывает наши данные, поэтому на основе этой прямой мы можем предсказывать и те данные, которых не было во входных.

**Например**: у нас есть данные о цене и площади квартир в каком-то районе. Корреляционная прямая может показать нам взаимосвязь площади квартиры с её ценой и мы, например, можем узнать, что чем больше площадь - тем выше цена. Но если бы нас попросили определить цену для какой-то квартиры - мы бы не смогли сделать это с помощью этой прямой. Тогда бы мы построили регрессионную прямую, которая показывает как площадь квартиры влияет на её стоимость. И если для какой-то квартиры (которой не было во входных данных) в этом районе нам известна площадь, а цена не известна, то мы можем использовать регрессионную прямую, чтобы "определить" цену на эту квартиру (с какой-то степенью точности).

В итоге, если мы хотим исследовать взаимосвязь двух количественных переменных.
- Мы можем построить scatter plot, чтобы убедиться, линейна ли взаимосвязь.
- Мы можем построить распределение наших остатков и посмотреть:
    - Нормально ли распределение остатков;
    - Ведут ли себя остатки одинаково на всем протяжении всей регрессионной прямой.
    
Если все условия выполняются, это означает, что наша модель очень хорошо описывает взаимосвязь переменных.

1. Строим регрессионную модель, которая объясняет взаимосвязь наших переменных. В нашем случае это будет уравнение регрессионной прямой: $$\widehat{y}=b_{0}+b_{1}x$$
Для того, чтобы эту модель построить, нам необходимо рассчитать коэффициент $b_0$ (intercept), который отвечает за то, где наша прямая пересекает ось $y$, и коэффициент $b_1$, который отвечает за угол наклона прямой.
2. Построив модель, мы хотим понять: насколько наша модель хорошо объясняет поведение двух переменных. Для этого найдем коэффициент детерминации $R^2$ и ответим на вопрос: какой % дисперсии объясняется нашей моделью. Наша нулевая гипотеза: 
$$H_0: b_1=0$$
3. Задача предсказания: как, основываясь на данных нашей НП можно предсказать ЗП.


```python
# Можно так решить
x=df["hs_grad"]
y=df["poverty"]
gradient, intercept, r_value, p_value, std_err = stats.linregress(x,y)
mn=np.min(x)
mx=np.max(x)
x1=np.linspace(mn,mx,500)
y1=gradient*x1+intercept

plt.plot(x,y,'ob') #Рассталяем точки
plt.plot(x1,y1,'-r') #Рисуем линию

plt.title("Связь бедности и уровня образования")
plt.xlabel("Уровень образования %")
plt.ylabel("Уровень бедности %")

plt.show() #Рисуем график
stats.linregress (x, y)
```

В полученной таблице выше выводятся: стандартная ошибка (std err), значение t-теста (t value) b p-уровень значимости($P>|t|$).  

P-уровень значимости в этом случае сильно меньше $p<0.05$, что позволяет нам отклонить нулевую гипотезу. Иными словами **мы обнаружили статистически значимую взаимосвязь между двумя нашими переменными**.

Так же можно обратить внимание, что и напротив Intercept-а рассчитана стандартная ошибка, значение t-теста и p-уровень значимости. И в этом случае $p<0.05$. Здесь так же проверяется гипотеза, что intercept отличен от нуля ($b_0 \ne 0$). В данном случае эффект не такой же значимый, как в предыдущем случае, но, всё-таки, **если предположить существование такого штата, где процент среднего образования равен нулю, то мы ожидаем что процент бедности в этом штате составит 64%**.

Убедившись, в том, что взаимосвязь между нашими переменными статистически значимая, проинтерпретируем само значение коэффициента $b_1$: **с каждым увеличением на 1% жителей в штате со средним образованием число бедности будет уменьшаться на 0.62%**.

Значение коэффициента детерминации (R-squared) равное 0.558 означает, что **примерно 56% изменчивости нашей зависимой переменной объясняется нашей моделью**. Это не так уж и много, потому что влияние многих факторов мы не учитываем.

# Регрессионный анализ с несколькими независимыми переменными

<center><img src="slides/week_3/week_3-28.png" title="slide 28" width="50%" height="50%"/></center>

**Множественная регрессия** (*Multiple regression*)позволяет исследовать влияние сразу нескольких независимых переменных (НП) на одну зависимую (ЗП).

Имея ЗП ($Y$) - "показатель бедности", обозначим влияние на нее фактора "уровень образования населения" (`rs_grad`), также фактора "процент белого населения в штате" (`white`)

\begin{equation*}
 \begin{cases}
   \widehat{y}=b_0+b_1X_1
   \\
   \widehat{y}=b_0+b_1X_1+b_2X_2
   \\
   ...
   \\
   \widehat{y}=b_0+b_1X_1+b_2X_2+...+b_nX_n
 \end{cases}
\end{equation*}

<center><img src="slides/week_3/week_3-29.png" title="slide 29" width="50%" height="50%"/></center>

При использовании, к примеру, 2 НП - делаем вывод, что оптимальным вариантом апроксимировать наши данные будет уже не прямая линия, а целая плоскость. Потому что мы уже работаем в трехмерном пространстве. Такая плоскость отображает наиболее успешный вариант описания взаимосвязи наших данных. Теперь **в качестве остатков будет выступать уже не расстояние до прямой линии, как в регрессии с одной НП, а расстояние до целой плоскости**. И мы также, методом наименьших квадратов (МНК) можем подобрать такие значения $b_0, b_1, b_2$, чтобы сумма всех этих ошибок, а именно квадратов расстояний (некоторые точки под плоскостью, некоторые над), была оптимальной по отношению ко всем точкам.

## Выбор модели

<center><img src="slides/week_3/week_3-33.png" title="slide 33" width="75%" height="75%"/></center>

- Сначала строим модель, в которую включаем все переменные (1 строка) и рассчитываем показатель "исправленный $R^2$". 

Мы будем по очериди удалять коэффициенты и смотреть влияние на показатель. **В регрессионном анализе количество не означает качество - если какая-либо переменная мешает, то лучше ее убрать**.

- Убираем `female_house`. Показатель $R^2=0.62$ улучшился.  
И так далее...

- Выбрав модель по лучшим показателям (без `female_house`), опять пробуем удалять различные переменные. В итоге, эта модель является наилучшей.

```python
'''
Программный перебор комбинаций НП
'''
# Найдём такой набор фичей, что R-squared максимальна
feats = {'metro_res', 'white', 'hs_grad', 'female_house'}

# Рекурсивно переберём все сочетания переменных
def brute_force(params: set, formula: str, result: set) -> str:
    if not params:
        result.add(formula.lstrip('+ '))
        return 
    else:
        new_params = params.copy()
        for el in params:
            new_params.remove(el)

            brute_force(new_params, formula + ' + ' + el, result)
            brute_force(new_params, formula, result)

res = set()          
brute_force(feats, '', res)
res.remove('')

# Теперь посчитаем Adjusted R-Square для каждого сочетания и выведем на экран
for formula in res:
    lm = ols(formula='poverty ~ ' + formula, data=df).fit()
    print(f'{lm.rsquared_adj:.2}', formula)
```

```python
# Function to calculate correlation coefficient between two arrays
def corr(x, y, **kwargs):
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.2, 0.45), size = 20, xycoords = ax.transAxes)

g = sns.PairGrid(df, diag_sharey=False)
# g.map_upper(sns.scatterplot)
g.map_upper(corr)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
```

Вернувшись к анализу, давайте посмотрим: а как, в принципе, взаимодействуют наши НП между собой. 
- Возьмем `white` и `hs_grade`. Эти НП очень незначительно связаны между собой и **это хороший показатель для модели. Когда каждая из НП взаимосвязана с ЗП, но при этом не сильно коррелирует между собой.**
- `metro_res` и `hs_grade` вообще практически не коррелируют между собой, но при этом по отдельности довольно хорошо связаны с нашей ЗП.
- `metro_res` и `white` ($\rho = -0.34$) коррелируют между собой, но мы ранее смотрели, что **включение обеих переменных в модель ситуацию не ухудшают**.

```python
# Итоговая модель
final_model = ols(formula='poverty ~ white + metro_res + hs_grad', data=df).fit()
final_model.summary()
```

Исключив из анализа переменную `female_house`, которая очень сильно коррелировала с остальными НП:
- Увеличили $R^2$;
- НП `white` так же стала значимо-взаимосвязана ($p=0.017$) с ЗП. Было $p=0.152$.

Наша НП `poverty` описывается моделью:
$$\widehat{y}=68.722-0.0533*\text{white}-0.0555*\text{metro_res}-0.5697*\text{hs_grad}$$

## Замечания
1. В регрессионном анализе (особенно во множественном) количество не всегда означает качество. Исключение переменных, которые сильно коррелируют с другими, только улучшит показатели модели.
2. "В чем приемущество этого метода по сравнению с корреляцией?" Мы же могли построить матрицу всех возможных корреляций и закончить наш анализ.
    - Корреляционные манипуляции - это ужасная история. Никогда их не используйте. Возникает опасность множественных сравнений. Возвращаемся к поправке Банферони и множественном сравнении в дисперсионном анализе.
3. Регрессионный анализ и его коэффициенты да/т нам более интересную информацию о взаимосвязи нашей ЗП и некоторого предиктора. Это не просто их корреляция. Это то, насколько они сильно связаны между собой, при условии, что мы учли все остальные переменные и они не изменяются. В случае регрессионного анализа мы не получим взаимосвязь, которая обусловлена третьей переменной.

## Логистическая регрессия

**Логистическая регрессия** или **логит-модель** (*logit model*) — это статистическая модель, используемая для прогнозирования вероятности возникновения некоторого события путём его сравнения с логистической кривой. Эта регреcсия выдаёт ответ в виде вероятности бинарного события (1 или 0).

### Пример

ЗП `admin` - успешность поступления абитуриента в ВУЗ.  
Предиктор (НЗ) `mark` - средняя оценка в школе.

Интерпретация здесь будет очень похожа, как и в случае линейной регрессии. Если наш коэффициент при НП ($1.0511$) статистически значимо отличаетя от нуля ($0.0004<0.05$), это означает, что с каждым единичным увеличением значения НП вероятность того, что абитуриент сдаст экзамен, будет увеличиваться.

Логистическая регрессия позволяет делать предсказания.

Задача кластеризации относится к статистической обработке, а также к широкому классу задач обучения без учителя.
